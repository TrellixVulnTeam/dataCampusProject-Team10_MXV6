## JSON에서 csv로 변환한 korquad 파일의 질문 정답 쌍을 summarization 데이터로 가공

````python
path = "/content/drive/Shared drives/BigDATA TEAM 10/dataset/korquadV2/Processed/"
csv_name = "konlpy comparison.csv"
````

````python
raw = pd.read_csv(os.path.join(path,csv_name))
````

````python
context = iter([_ for _ in raw.context])
nouns = iter([_ for _ in raw.komoran])


collect_set = []
collect_context = []

cur_context = next(context)

i = 0
j = 0
while i<len(raw):
  a = True
  current_set = set()

  while a:
    try: 
      curr = next(nouns)[1:-1]
      curr = curr.replace("' ", '').replace("'",'')
      temp = curr.split(",")
      current_set.update(temp)

      print(i)
      i+=1

      old_context = cur_context
      cur_context = next(context)
      
      if cur_context!=old_context:
        print("context differ")
        j += 1
        a = False
        collect_context.append(old_context)
        collect_set.append(current_set)
    except StopIteration:
      break

  
````

````python
print(j)
print(collect_set[:2])
````

````python
summary_dicts = dict(zip(collect_context, collect_set))
````

````python
df = pd.DataFrame({'context':collect_context,'nouns':collect_set})
````
